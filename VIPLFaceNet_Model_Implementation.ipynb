```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D, ZeroPadding2D
from keras.layers.normalization import BatchNormalization
from keras.regularizers import l2

def VIPLFaceNet_model(img_shape=(224, 224, 3), n_classes=10, l2_reg=0.,
    weights=None):

    # Initialize model
    VIPLFaceNet = Sequential()

    # Layer 1
    #Change from (11,11) to (9,9), BatchNorm remove
    VIPLFaceNet.add(Conv2D(96, (9, 9), input_shape=img_shape,
        strides=(4, 4),padding='same', kernel_regularizer=l2(l2_reg)))
    VIPLFaceNet.add(Activation('relu'))
    VIPLFaceNet.add(MaxPooling2D(pool_size=(2, 2)))

    # Layer 2 
    #Change (5, 5) to (3, 3)
    VIPLFaceNet.add(ZeroPadding2D((1, 1)))
    VIPLFaceNet.add(Conv2D(192, (3, 3), padding='same'))
    VIPLFaceNet.add(Activation('relu'))

    # Layer 3 
    #Change (5, 5) to (3, 3) 
    VIPLFaceNet.add(ZeroPadding2D((1, 1)))
    VIPLFaceNet.add(Conv2D(192, (3, 3), padding='same'))
    VIPLFaceNet.add(Activation('relu'))
    VIPLFaceNet.add(MaxPooling2D(pool_size=(2, 2)))

    # Layer 4
    VIPLFaceNet.add(ZeroPadding2D((1, 1)))
    VIPLFaceNet.add(Conv2D(384, (3, 3), padding='same'))
    VIPLFaceNet.add(Activation('relu'))

    # Layer 5
    VIPLFaceNet.add(ZeroPadding2D((1, 1)))
    VIPLFaceNet.add(Conv2D(256, (3, 3), padding='same'))
    VIPLFaceNet.add(Activation('relu'))

    # Layer 6
    VIPLFaceNet.add(ZeroPadding2D((1, 1)))
    VIPLFaceNet.add(Conv2D(256, (3, 3), padding='same'))
    VIPLFaceNet.add(Activation('relu'))

    # Layer 7
    VIPLFaceNet.add(ZeroPadding2D((1, 1)))
    VIPLFaceNet.add(Conv2D(192, (3, 3), padding='same'))
    VIPLFaceNet.add(Activation('relu'))
    VIPLFaceNet.add(MaxPooling2D(pool_size=(2, 2)))
    
    # Layer 8
    VIPLFaceNet.add(Flatten())
    VIPLFaceNet.add(Dense(4096))
    VIPLFaceNet.add(Activation('relu'))
    VIPLFaceNet.add(Dropout(0.5))

    # Layer 9
    #Change 4096 to 2048
    VIPLFaceNet.add(Dense(2048))
    VIPLFaceNet.add(Activation('relu'))
    VIPLFaceNet.add(Dropout(0.5))

    # Layer 10
    VIPLFaceNet.add(Dense(10575))
    VIPLFaceNet.add(Activation('softmax'))

    if weights is not None:
        VIPLFaceNet.load_weights(weights)

    return VIPLFaceNet
```
